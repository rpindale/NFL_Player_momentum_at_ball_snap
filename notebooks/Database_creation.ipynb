```python
# Create the database using sqlite3
BDB25_db = sqlite3.connect("big_data_bowl_2025.db") 

# If getting error in adding tables, it is likely that the database file is not writable. Solve by using below. 
#! chmod 777 big_data_bowl_2025.db
```

```python

# Lists the names of the csv files provided in the competetion.
table_names = [
    'plays',
    'tracking_week_5',
    'tracking_week_4',
    'tracking_week_6',
    'tracking_week_7',
    'tracking_week_3',
    'tracking_week_2',
    'tracking_week_1',
    'players',
    'games',
    'tracking_week_9',
    'tracking_week_8',
    'player_play'
]

for table_name in table_names:
    
    # Reads the csv files
    table = pd.read_csv(f"nfl-big-data-bowl-2025/{table_name}.csv")
    
    # Create tables in the database
    table.to_sql(
        table_name, 
        BDB25_db, 
        index=False, 
        chunksize=1000, 
        if_exists='replace'
    )
```

```python
# Create a cursor so that we can query the database
BDB25_cursor = BDB25_db.cursor()
```

```python
BDB25_db.commit()
```

```python
BDB25_db.close()
```
